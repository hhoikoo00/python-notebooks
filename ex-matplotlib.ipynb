# Matplotlib Exercise

## Setup

```python
import numpy as np
import matplotlib.pyplot as plt

resolution = 1000
x = np.linspace(0, 50, resolution)
y = np.sin(x)
z = np.exp(y)
```

## Questions
1. Plot y against x and call this figure 1.
2. Add a title to this graph, call this figure 2, and make the title 'My Graph'
3. Add x and y labels to your graph, call this figure 3, make the y axis called ‘Amplitude (W)’ and the x axis called ‘time (s)’
4. Plot y AND z against x on the same graph
5. Make the y plot the colour ‘Blue’ and the z plot the colour ‘Red’
6. Add a legend onto the graph to make sure you can tell which is y and z
7. Restrict the y scale to a range of -0.5 to 2.0
8. Save the file as a high resolution .png file

```python
# Question 1
plt.figure(1)
y_line, = plt.plot(x, y, label='y', color='blue') # Question 5

# Question 2
plt.title('My Graph')

# Question 3
plt.ylabel('Amplitude (W)')
plt.xlabel('time (s)')

# Question 4
# y plotted above
z_line, = plt.plot(x, z, label='z', color='red') # Question 5

# Question 6
plt.legend(handles=[y_line, z_line])

# Question 7
plt.ylim(ymin=-0.5, ymax=2.0)

# Question 8
plt.savefig('foo.png', bbox_inches='tight')

# Show the graph
plt.show()
```
